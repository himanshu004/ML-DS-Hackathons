```
from pycaret.classification import *
import pandas as pd
import numpy as np
```

```
train = pd.read_csv('train.csv',index_col = 'PassengerId')
clf1 = setup(train,target = 'Survived',ignore_features = ['Name','Ticket'])
```

```
compare_models(whitelist = ['rf','dt','xgboost'],sort = 'Accuracy',)
```

```
xgb = create_model(estimator = 'xgboost',ensemble = True,method = 'Boosting')
```

```
tuned_xgb = tune_model(estimator = xgb,optimize = 'Accuracy',choose_better = True)
```

```
evaluate_model(tuned_xgb)
```

```
test = pd.read_csv('test.csv',index_col = 'PassengerId' )
result = test[['Age','Name']]
result
```

```
predictions = predict_model(tuned_xgb,data = test)
predictions['Label']
```

```
result['Survived'] = predictions['Label'].values
result.drop(['Age','Name'],axis = 1,inplace = True)
result
```

```
result.to_csv('pycaret_submission.csv')
```
